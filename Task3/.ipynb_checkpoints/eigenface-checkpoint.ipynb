```python
import os
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
from matplotlib.pyplot import imread
```

```python
# calculate the mean face and normalized faces
def preprocess(A_pp):
    A = None
    Q_norms = None
    A_means = None


    A_pp_mean = np.mean(A_pp, axis=1, keepdims=True)
    Q = A_pp - A_pp_mean
    Q_norms = np.amax(np.abs(Q), axis=1, keepdims=True)
    A = Q / np.where(Q_norms == 0, 1, Q_norms)
    A_means = np.mean(A_pp, axis=1, keepdims=True)

    return A, Q_norms, A_means


# the main algorithm of eigenface.
def eigen_faces(A_pp):
    C = None
    F = None
    D = None
    Q_norms = None
    A_means = None


    A, Q_norms, A_means = preprocess(A_pp)
    # D: eigenvalues C: their corresponding symmetric matrix for A.T @ A
    # where A.T @ A is the covariance matrix of the original faces set
    # which is a faster way of computing eigenface than directly using A @ A.T.
    D, C = np.linalg.eig(A.T @ A)
    
    # make real number in matrix
    D = D.real
    C = C.real

    # eigenfaces
    F = A @ C
    F_norm = np.linalg.norm(F, axis=0, keepdims=True)
    # normalized faces
    F = F / F_norm

    # compute
    C = A.T @ F 
    return A, C, F, D, Q_norms, A_means
```

```python
def reconstruct_face(Img, F, Q_norms, A_means,A):
    R = np.zeros((231, 195))


    #     calculate C, which is AT@A's eigenvector matrix
    C = A.T @ F

    #     calculate R_vector
    R_vector = np.zeros(Q_norms.shape)

    for j in range(len(Img)):
        R_vector = R_vector + (Img[j] * Q_norms * F[:, j].reshape(-1, 1))

    R_vector += A_means
    R = R_vector.reshape((231, 195), order="F")

    return R


# top k faces projection
def face_projection(image_vector, k, F, D, A_means, Q_norms):
    compressed_image = None
    p = None
    #     k: top k int
    #     F:eigen Face， A@A.T   eigenvector 45045,135
    #     D: A.T@A eigenvalue 135,
    #     A_means: mean face  45045,1
    #     Q_norms: image normalized       45045,1
    #     image_vector: image we need to compress ,45045,1

    #     intialize image_vector
    num_images = D.shape[0]
    Q = image_vector.reshape(-1, 1) - A_means
    image_normalized = Q / np.where(Q_norms == 0, 1, Q_norms)

    #     handle compressed image
    F_k = F[:, 0:k]

    compressed_parameter = F_k.T @ image_normalized
    compressed_parameter_s = np.zeros((num_images, 1))
    for i in range(num_images):
        if (i < k):
            compressed_parameter_s[i] = compressed_parameter[i]

    #     handle proportion
    d_sum = np.sum(D)
    numerator = 0
    for i in range(k):
        numerator += D[i]
    proportion = numerator / d_sum

#     print(image_normalized.shape)
#     print(compressed_parameter.shape)
#     print(F_k.shape)


    return compressed_parameter_s[:k], proportion
```

```python
# find k nearest neighbour among all training faces, sort them by the Euclidean distance.
def k_nearest_face(test_face,training_face_params):
    all_distances = []
    for i,face in enumerate(training_face_params):
        dis = (i,np.linalg.norm((face-test_face), ord=2, axis=None, keepdims=False))
        all_distances.append(dis)
    
    return sorted(all_distances,key=lambda x:x[1])
```

```python
# k nearest neighbour
nearest = 3
ans_faces = k_nearest_face(params_test_face,training_face_params)
print(np.array(ans_faces[:nearest]))
```

```python
# plotting nearest face
plt.figure(num='nearest face',figsize=(12,12))

plt.subplot(1,3,1)
plt.title(f"the nearest face")
plt.imshow(yale_images[ans_faces[0][0]].reshape((231, 195),order='F'),cmap='gray')


plt.subplot(1,3,2)
plt.title(f"the second nearest")
plt.imshow(yale_images[ans_faces[1][0]].reshape((231, 195),order='F'),cmap='gray')


plt.subplot(1,3,3)
plt.title(f"the third nearest")
plt.imshow(yale_images[ans_faces[2][0]].reshape((231, 195),order='F'),cmap='gray')


plt.show()
```
